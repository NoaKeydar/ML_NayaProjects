```python
class Node():
    # name = the name of the node
    # pointing at={1:10,2:15..}

    def __init__(self, name):
        self.name = name
        self.pointing_at = {}
        self.count_neigh = 0

    def __str__(self):
        return  str(self.name)

    def __len__(self):
        value = len(self.pointing_at.keys())
        return int(value)

    def __eq__(self, other):
        if self.name == str(other):
            return 'equal'

    def __ne__(self, other):
        if self.name != str(other):
            return 'not equal'

    def is_pointing_at(self, name):
        if name in self.pointing_at.keys():
            return True
        else:
            return False

    def add_neighbor(self, name, weight=1):
        if self.name not in self.pointing_at.keys():
            self.pointing_at[name] = weight
        return str(self.pointing_at)

    def remove_neighbor(self, name):
        del self.pointing_at[name]

    def get_weight(self, name):
        if name in self.pointing_at.keys():
            return self.pointing_at[name]
        else:
            return 0
```

```python

#The Graph Class

class Graph:
# name = the name of the graph
# nodes = keys are nodes name, value are nodes instance Node(1)=node_list[0]

    def __init__(self,name,nodes=[]):
        self.name = name
        self.nodes = {}
        self.edges = {}
        self.num_nodes = 0

        for n in nodes:
            if type(n) == int:
                self.nodes[n] = list_node[n-1]
                self.edges[n] = list_node[n-1].pointing_at
            if type(n) == str:
                for k in destination_dict.keys():
                    if k==n:
                        self.nodes[k] = destination_dict[k]
                        self.edges[k] = destination_dict[k]

    def __str__(self):
        return 'Graph = ' + self.name+ "," + " Nodes = " + str(self.nodes.keys())

    def __len__(self):
        value = len(self.nodes.keys())
        return int(value)

    def __contains__(self, key):
        if (isinstance(key, str)) and (key in self.nodes.keys()):
            val = True
        if isinstance(key,Node) and (key in self.nodes.values()):
            val = True
        val = False
        return val

    def __getitem__(self,name):
        if name in self.nodes.keys():
            val = Node(name)
        else:
            val = 'This Node is not in the Graph'
        return val

    def add(self,other):
        new_graph = Graph('new', [])
        for i in self.nodes:
            new_graph.add_node(self.nodes[i])
        for num in other.nodes.keys():
            new_graph.add_node(other.nodes[num])

        return new_graph

    def add_node(self,node):
        ins_name = node.name
        
        if type(ins_name) == int:
            if ins_name not in self.nodes.keys():
                    self.nodes[ins_name] = node
                    self.edges[ins_name] = list_node[int(ins_name) -1].pointing_at
            elif ins_name in self.nodes.keys():
                    self.edges[ins_name] = list_node[int(ins_name)-1].pointing_at
       
        if type(ins_name) == str:
            if ins_name not in self.nodes.keys():
                    self.nodes[ins_name] = destination_dict[ins_name]
                    self.edges[ins_name] = destination_dict[ins_name].pointing_at
            elif ins_name in self.nodes.keys():
                    self.edges[ins_name] = destination_dict[ins_name].pointing_at
                
        return self.edges
    
    def remove_node(self,name):
        del self.nodes[name]

    def is_edge(self,frm_name,to_name):
        if to_name in list_node[frm_name-1].pointing_at.keys():
            return True
        return False


    def add_edge(self,frm_name,to_name,weight=1):
        new_edge={}
        if frm_name not in self.nodes:
            self.add_node(Node(frm_name))

        if to_name not in self.edges[frm_name]:
            new_edge[to_name] = weight
            self.edges[frm_name].update(new_edge)

            return self.edges


    def remove_edge(self,frm_name,to_name):
        del self.edges[frm_name][to_name]
        return self.edges

    def get_edge_weight(self,frm_name,to_name):
        if self.is_edge(frm_name,to_name) == True:
            return self.edges[frm_name][to_name]

        else:
            return 'none'

    def get_path_weight(self,path):
        sum = 0
        i=0

        for p in range(len(path)-1):
            if path[p+1] not in self.edges[path[p]]:
                return "Invalid path"

        while i+1 < len(path):
            sum = sum + self.get_edge_weight(path[i],path[i+1])
            i=i+1


        if sum ==0:
            return 'None'
        if sum >0:
            return sum
# returns all the possible paths

    def find_all_paths(self,frm_name,to_name,path=[]):
        path = path + [frm_name]
        paths = []
        if frm_name == to_name:
            return [path]
        if not frm_name in self.nodes.keys():
            return []
        for i in self.edges[frm_name].keys():
            if i not in path:
                newpaths = self.find_all_paths(i, to_name, path)
                for newpath in newpaths:
                    paths.append(newpath)
        return paths


    def is_reachable(self,frm_name,to_name):
        if not self.find_all_paths(frm_name, to_name) or frm_name==to_name:
            return False
        else:
            return True


    def is_isolated(self):
        self.isolist =[]
        for i in range(len(self.edges.keys())):
            self.isolist.extend(g4.edges.values()[i].keys())

        for var in self.nodes:
            if var not in self.isolist:
                return str(var) + ' Is isolated'


    def find_shortest_path(self,frm_name,to_name,min_weight=[]):
        self.list_paths = self.find_all_paths(frm_name,to_name)
        self.Max_weight = 50 #set threshold
        self.path = []


        if not frm_name in self.nodes.keys() or not to_name in self.nodes.keys():
                return "Not valid frm_name/to_name"

        for i in self.list_paths:
                weight = self.get_path_weight(i)
                if weight < self.Max_weight:
                    self.Max_weight=weight
                    self.min_weight =weight
                    self.path = i

        return   self.min_weight ,self.path

```

```python
import time
from datetime import datetime
from datetime import timedelta
import csv

Travels_EW = r"travelsEW.csv"
Travels_WE=r"travelsWE.csv"

clean_data=[]
clean_dataWE=[]
#Travels_EW

with open(Travels_EW, 'r') as f:
    data=csv.reader(f)

    #remove header
    next(data)

    #cleaning the data
    for line in data:
        if (line[0]=='Center' or  line[0]=='East') and (line[2] == 'West' or line[2] == 'North' or line[2] == 'South' ) and ( line[1] != 0 ) and (line[3] != 0) \
                and (line[1]!="26/02/2016 1412") and (line[3] != "22/01/2016 25h52m"):
            dt1 = datetime.strptime(line[1], "%d/%m/%Y %Hh%Mm")
            dt2 = datetime.strptime(line[3], "%d/%m/%Y %Hh%Mm")
            t1 = datetime(dt1.year, dt1.month, dt1.day, dt1.hour, dt1.minute)
            t2 = datetime(dt2.year, dt2.month, dt2.day, dt2.hour, dt2.minute)
            line[1] = time.mktime(t1.timetuple())
            line[3] = time.mktime(t2.timetuple())
            line.append(timedelta.total_seconds(t2-t1))
            clean_data.append(line)



#distinct values from departures and arivals columns
departure=[]
arrival=[]
for row in clean_data:
    departure.append(row[0])
    arrival.append(row[2])

departure_distinct=set(departure)
arrival_distinct=set(arrival)

#calc average time per destination
averageEW=[]
for v in departure_distinct:
    for k in arrival_distinct:
        count = 0
        sum = 0
        for row in clean_data:
            if row[0] == v and row[2] == k:
                count=count+1
                sum =row[3]+sum

        if count > 0:
            x=int(sum/count)
            averageEW.extend((k,v,x))


print(averageEW)
```
